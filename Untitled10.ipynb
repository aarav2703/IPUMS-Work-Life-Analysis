```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score


X = df[['HOURSWRK']]
y = df['HEALTH']


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


model = LinearRegression()


model.fit(X_train, y_train)


y_pred = model.predict(X_test)


mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse}')
print(f'R-squared: {r2}')


plt.figure(figsize=(10, 6))
plt.scatter(X_test, y_test, color='blue', label='Actual Health Status')
plt.plot(X_test, y_pred, color='red', linewidth=2, label='Predicted Health Status')
plt.title('Linear Regression: Health Status vs. Hours Worked')
plt.xlabel('Hours Worked Per Week')
plt.ylabel('Health Status')
plt.legend()
plt.show()
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix


X = df[['EMPSTAT']]
y = df['DEPFREQ'].apply(lambda x: 1 if x == 1 else 0)


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)


log_model = LogisticRegression()


log_model.fit(X_train, y_train)


y_pred = log_model.predict(X_test)


print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))


plt.figure(figsize=(10, 6))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='coolwarm', xticklabels=['No Depression', 'Daily Depression'], yticklabels=['No Depression', 'Daily Depression'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix: Depression Frequency vs. Employment Status')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from sklearn.linear_model import LinearRegression, LogisticRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, classification_report, confusion_matrix
import zipfile
import os


os.makedirs('visualizations', exist_ok=True)


plt.figure(figsize=(10, 6))
sns.histplot(df['HOURSWRK'], bins=30, kde=False)
plt.title('Distribution of Hours Worked')
plt.xlabel('Hours Worked Per Week')
plt.ylabel('Frequency')
plt.savefig('visualizations/hours_worked_distribution.png')


plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='HEALTH', order=df['HEALTH'].value_counts().index)
plt.title('Distribution of Health Status')
plt.xlabel('Health Status')
plt.ylabel('Count')
plt.savefig('visualizations/health_status_distribution.png')


plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='EMPSTAT', y='HEALTH', ci=None)
plt.title('Average Health Status by Employment Status')
plt.xlabel('Employment Status')
plt.ylabel('Average Health Status')
plt.savefig('visualizations/health_by_employment_status.png')


plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='DEPFREQ', order=df['DEPFREQ'].value_counts().index)
plt.title('Distribution of Depression Frequency')
plt.xlabel('Depression Frequency')
plt.ylabel('Count')
plt.savefig('visualizations/depression_frequency_distribution.png')


numeric_df = df.select_dtypes(include=['int64', 'float64'])
corr_matrix = numeric_df.corr()
plt.figure(figsize=(14, 10))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm')
plt.title('Correlation Matrix of Selected Variables')
plt.savefig('visualizations/correlation_matrix.png')


X = df[['HOURSWRK']]
y = df['HEALTH']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
model = LinearRegression()
model.fit(X_train, y_train)
y_pred = model.predict(X_test)
plt.figure(figsize=(10, 6))
plt.scatter(X_test, y_test, color='blue', label='Actual Health Status')
plt.plot(X_test, y_pred, color='red', linewidth=2, label='Predicted Health Status')
plt.title('Linear Regression: Health Status vs. Hours Worked')
plt.xlabel('Hours Worked Per Week')
plt.ylabel('Health Status')
plt.legend()
plt.savefig('visualizations/linear_regression_health_hours.png')


X = df[['EMPSTAT']]
y = df['DEPFREQ'].apply(lambda x: 1 if x == 1 else 0)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
log_model = LogisticRegression()
log_model.fit(X_train, y_train)
y_pred = log_model.predict(X_test)
plt.figure(figsize=(10, 6))
sns.heatmap(confusion_matrix(y_test, y_pred), annot=True, fmt='d', cmap='coolwarm', xticklabels=['No Depression', 'Daily Depression'], yticklabels=['No Depression', 'Daily Depression'])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix: Depression Frequency vs. Employment Status')
plt.savefig('visualizations/logistic_regression_confusion_matrix.png')


with zipfile.ZipFile('visualizations.zip', 'w') as zipf:
    for root, _, files in os.walk('visualizations'):
        for file in files:
            zipf.write(os.path.join(root, file), arcname=file)


os.listdir()
```
